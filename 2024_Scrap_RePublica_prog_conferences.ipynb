#Scrapping avec beautiful soup

##installation

```python
import requests, time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#url = 'https://re-publica.com/de/schedule?day=2024-05-27'
url = 'https://re-publica.com/de/schedule?day=2024-05-28'
#url = 'https://re-publica.com/de/schedule?day=2024-05-29'


navigator = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1)'
html = requests.get(url, headers={'User-Agent': navigator})

html
```

##parsing

```python
soup = BeautifulSoup(html.content, 'html.parser')
```

```python
#dans le containeur général, on veut identifier les containeurs de chaque conférence
# ces dernières ont des div qui cumulent deux classes : session et scheduled
# on peut les récupérer en interrogeant ces attributs
items=[c for c in soup.find_all(attrs=({'class': ['session',"scheduled"]}))]
print(items[0])
```

```python
len(items)
```

```python
#définir le containeur général de chaque conference
conference = soup.find(id='rp-vertical')
print(conference)
```

les containers :

1.   horaire début class="start"
2.   horaire fin class="end"

*   class="session-title node__title"
*   class="speakers"

*  class="description"
*   class="stage"

*   class="language"
*   class="session-type"








##pour récupérer les infos de la première conférence du jour

```python
#d'abord le titre de la conf
items[0].find(class_='session-title node__title')
```

```python
#pour récupérer le texte de la conf seulement et pas toute la ligne de code
print(items[0].find(class_='session-title node__title').get_text())

#pour les intervenants
print(items[0].find(class_='speakers').get_text())

#pour la description de la conf
print(items[0].find(class_='description').get_text())

#pour la description de la scène
print(items[0].find(class_='stage').get_text())

#pour la description de la langue
print(items[0].find(class_='language').get_text())

#pour la description de le type de session
print(items[0].find(class_='session-type').get_text())

#pour la description de l'heure de début
print(items[0].find(class_='start').get_text())
#pour la description de l'heure de fin
print(items[0].find(class_='end').get_text())
```

## pour récupérer les infos de toutes les conf du jour via une boucle et les mettre dans une liste

```python
#pour créer une boucle qui récupère - dans une liste - ces infos pour chaque jour
session_names = [item.find(class_='session-title node__title').get_text() for item in items]
intervenants = [item.find(class_='speakers').get_text() for item in items]
description = [item.find(class_='description').get_text() if item.find(class_='temp') else None for item in items ]
scenes = [item.find(class_='stage').get_text() for item in items]
langue = [item.find(class_='language').get_text() for item in items]
session_type = [item.find(class_='session-type').get_text() for item in items]
heure_debut = [item.find(class_='start').get_text() for item in items]
heure_fin = [item.find(class_='end').get_text() for item in items]
theme = [item.find(class_='track').get_text() for item in items]


#attention si case vide, je lui dis de remplir avec None avec le if... else
print(session_names)
print(intervenants)
print(description)
print(scenes)
print(langue)
print(session_type)
print(heure_debut)
print(heure_fin)
print(theme)
```

##en faire un dataframe

```python
#on utilise un dictionnaire

#conferences_day1 = pd.DataFrame(
conferences_day2 = pd.DataFrame(
#conferences_day3 = pd.DataFrame(
    {
        'intitulé': session_names, #nom de la colonne:valeur à intégrer
        'intervenants': intervenants,
        'description': description,
        'n° de scène': scenes,
        'langue': langue,
        'type de session': session_type,
        'heure de début': heure_debut,
        'heure de fin': heure_fin,
     })
#print(conferences_day1)
print(conferences_day2)
#print(conferences_day3)
```
