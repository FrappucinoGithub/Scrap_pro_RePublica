## séparer les noms dans intervenants en plusieurs

Quand je veux séparer les noms à partir de la virgule en créant des colonnes pour chaque nom, j'ai un message d'erreur : "ValueError: Columns must be same length as key". Je contourne en faisant d'abord une séparation des noms dans un dataframe temporaire

```python
tmp = pd.concat( [prog3days['intervenants'].str.split(',', expand=True).add_prefix('intervenant')], axis=1 )
tmp
```

```python
tmp = tmp[['SubColumn0', 'SubColumn1', 'SubColumn2', 'SubColumn3', 'SubColumn4', 'SubColumn5', 'SubColumn6', 'SubColumn7']]
tmp
```

```python
prog_concat = pd.concat([prog3days, tmp], axis=1)
prog_concat=prog_concat.drop(columns=['intervenants'])
prog_concat
```
